```
train_data['data']='train'
test_data['data']='test'
data = pd.concat([train_data,test_data],axis=0, sort=True)
data.head()
```

```
data.drop(["Customer Id", "Artist Name", "Artist Reputation"],axis=1,inplace=True)
data
```

```
data["Cost"] = data["Cost"].abs()
```

```
data["Height"].fillna(train_data['Height'].median(), inplace=True)
data["Weight"].fillna(train_data['Weight'].median(), inplace=True)
data["Width"].fillna(train_data['Width'].median(), inplace=True)
```

```
data["Transport"].fillna(train_data['Transport'].mode()[0], inplace=True)
```

```
data.isna().sum()
```

```
train_data.isna().sum()
```

```
data = data.replace({'No': 0, 'Yes': 1})
```

```
bool_cols = ["International", "Express Shipment", "Installation Included", "Remote Location", "Fragile"]
for col in bool_cols:
  data[col] = pd.to_numeric(data[col])
data.info()
```

```
train_data = data[data['data']=='train']
del train_data['data']
train_data.columns
```

```
test_data = data[data['data']=='test']
del test_data['data']
del test_data['Cost']
test_data.columns
```

```
train_data.dropna(inplace=True)
```

```
from sklearn.preprocessing import OneHotEncoder

cat_cols = ["Customer Information", "Material", "Transport"]
enc = OneHotEncoder(handle_unknown='ignore', sparse=False)
enc.fit(train_data[cat_cols])
train_encoded = enc.transform(train_data[cat_cols])
test_encoded = enc.transform(test_data[cat_cols])
```

```
encoded_cols = [cat for col in enc.categories_ for cat in col]
encoded_cols
```

```
train_data.join(pd.DataFrame(train_encoded, columns=encoded_cols, index=train_data.index))

test_data.join(pd.DataFrame(test_encoded, columns=encoded_cols, index=test_data.index))
```

```
train_data.drop(cat_cols,axis=1,inplace=True)
test_data.drop(cat_cols,axis=1,inplace=True)
```

```
train_data.info()
```

```
train_data["Delivery Date"] = pd.to_datetime(train_data["Delivery Date"])
test_data["Delivery Date"] = pd.to_datetime(test_data["Delivery Date"])
# train_data["Scheduled Date"] = pd.to_datetime(train_data["Scheduled Date"])
```

```
# Extract features - year, month, day, hour
train_data['year'] = train_data['Delivery Date'].dt.year 
train_data['month'] = train_data['Delivery Date'].dt.month 
train_data['day'] = train_data['Delivery Date'].dt.day

# Extract features - year, month, day, hour
test_data['year'] = test_data['Delivery Date'].dt.year 
test_data['month'] = test_data['Delivery Date'].dt.month 
test_data['day'] = test_data['Delivery Date'].dt.day
```

```
# train_data["date_diff"] = (train_data["Delivery Date"] - train_data["Scheduled Date"]).map(lambda x:str(x).split()[0])
# train_data['date_diff'] = pd.to_numeric(train_data['date_diff'])
```

```
train_data.drop(["Delivery Date", "Scheduled Date"],axis=1,inplace=True)
test_data.drop(["Delivery Date", "Scheduled Date"],axis=1,inplace=True)
```

```
# train_data['pincode'] = train_data['Customer Location'].map(lambda x:x.split()[-1])
# train_data['state'] = train_data['Customer Location'].map(lambda x:x.split()[-2])
# train_data['city'] = train_data['Customer Location'].map(lambda x:" ".join(x.split()[:-2])[:-1])
train_data.drop('Customer Location', inplace=True, axis=1)
test_data.drop('Customer Location', inplace=True, axis=1)
```
